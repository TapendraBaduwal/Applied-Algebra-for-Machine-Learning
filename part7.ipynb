```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC
import numpy as np

path_dataset = '/content/train_dataset1.csv'
train_dataset = pd.read_csv(path_dataset)

print("Dataset:")
print(train_dataset.head())

X = train_dataset.iloc[:, :-1].values
y = train_dataset.iloc[:, -1].values

# Plot original data to check for linear separability
def plot_data(X, y, title):
    plt.scatter(X[:, 0], X[:, 1], c=y, cmap=plt.cm.coolwarm, edgecolors='k')
    plt.title(title)
    plt.xlabel('Feature 1')
    plt.ylabel('Feature 2')
    plt.show()

plot_data(X, y, "Scatter Plot of Original Data Points")

# Standardize the data
scaler = StandardScaler()
X = scaler.fit_transform(X)

# Plot modified data after standardization
plot_data(X, y, "Scatter Plot of Standardized Data Points")

# Function to plot decision boundary
def plot_decision_boundary(clf, X, y, title):
    h = .02  # step size in the mesh
    x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
    y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
    xx, yy = np.meshgrid(np.arange(x_min, x_max, h),
                         np.arange(y_min, y_max, h))

    Z = clf.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)

    plt.contourf(xx, yy, Z, alpha=0.8)
    plt.scatter(X[:, 0], X[:, 1], c=y, edgecolors='k', marker='o')
    plt.title(title)
    plt.xlabel('Feature 1')
    plt.ylabel('Feature 2')
    plt.show()

# Train linear SVM and plot decision boundary (SVM without kernel trick)
clf_linear = SVC(kernel='linear')
clf_linear.fit(X, y)
plot_decision_boundary(clf_linear, X, y, "Linear Kernel Decision Boundary")

# Train SVM with RBF kernel and plot decision boundary
clf_rbf = SVC(kernel='rbf')
clf_rbf.fit(X, y)
plot_decision_boundary(clf_rbf, X, y, "RBF Kernel Decision Boundary")

# Train SVM with Polynomial kernel and plot decision boundary
clf_poly = SVC(kernel='poly', degree=3)
clf_poly.fit(X, y)
plot_decision_boundary(clf_poly, X, y, "Polynomial Kernel Decision Boundary")
```
